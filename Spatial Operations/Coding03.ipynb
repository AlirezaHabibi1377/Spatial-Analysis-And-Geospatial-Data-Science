```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = gpd.read_file("data/countries.shp")
countries.head()
```

```python
countries_centroid = countries.geometry.centroid
countries_centroid.head()
```

```python
fig,ax = plt.subplots(figsize=(15,10))

countries.plot(ax = ax)
countries_centroid.plot(ax = ax, color = 'red')
```

```python
Chile = countries[countries['NAME'] == 'Chile']

fig,ax = plt.subplots(figsize=(15,10))

Chile.plot(ax = ax)
Chile_centroid.plot(ax = ax, color = 'red')

Chile.crs

Chile_centroid = Chile.geometry.centroid
```

```python
Chile_buffer_1dg = Chile_centroid.buffer(1)
Chile_buffer_2dg = Chile_centroid.buffer(2)

fig,ax = plt.subplots(figsize=(15,10))
Chile.plot(ax = ax)
Chile_buffer_2dg.plot(ax = ax, color = 'yellow')
Chile_buffer_1dg.plot(ax = ax, color = 'red')
```
